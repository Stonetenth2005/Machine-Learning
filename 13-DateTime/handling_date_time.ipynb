```python
import numpy as np
import pandas as pd
```

```python
date = pd.read_csv('/content/sales_data.csv')
time = pd.read_csv('/content/messages.csv', encoding='latin1')
```

```python
df_dict = {
    'date': date['Order Date'],
}

date = pd.DataFrame(df_dict)
date.head()
```

#  Working with Dates

```python
# Converting to date-time

date['date'] = pd.to_datetime(date['date'])
```

Extracting Year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

Extracting Month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

Extracting Day

```python
date['day'] = date['date'].dt.day
date.head()
```

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head()
```

```python
date['date_name'] = date['date'].dt.day_name()
date.head()
```

```python
# is weekend ?

date['date_is_weekend'] = np.where(date['date_name'].isin(['Sunday', 'Saturday']), 1, 0)
date.head()
```

Extracting Week

```python
date['date_week'] = date['date'].dt.isocalendar().week
date.head()
```

Extracting Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

Extract Semester

```python
date['semester'] = np.where(date['quarter'].isin([1, 2]), 1, 2)
date.head()
```

Extract Time elapsed between dates

```python
import datetime
today = datetime.datetime.today()
today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
# Months passed
months_passed = (today.year - date['date'].dt.year) * 12 + (today.month - date['date'].dt.month)
np.round(months_passed, 0)
```

# Working on Time

```python
time['date'] = pd.to_datetime(time['date'])
time.info()
```

```python
time['hour'] = time['date'].dt.hour
time['min'] = time['date'].dt.minute
time['sec'] = time['date'].dt.second

time.head()
```

Extract Time part

```python
time['time'] = time['date'].dt.time

time.head()
```

Time difference

```python
today - time['date']
```

```python
# in seconds

(today - time['date'])/np.timedelta64(1, 's')
```

```python
# in minutes

(today - time['date'])/np.timedelta64(1, 'm')
```

```python
# in hours

(today - time['date'])/np.timedelta64(1, 'h')
```
