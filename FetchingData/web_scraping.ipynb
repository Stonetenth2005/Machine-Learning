```python
import pandas as pd
import requests
from bs4 import BeautifulSoup # web scraping library
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/117.0"
}
```

```python
df_final = pd.DataFrame()
```

```python
# Running for 15 pages

for i in range(1,15):
  url = 'https://books.toscrape.com/catalogue/page-{i}.html'
  webPage = requests.get(url, headers=headers).text

  soup = BeautifulSoup(webPage, 'html.parser')

  name = []
  rating = []
  price = []
  is_in_stock = []

  books = soup.find_all("article", class_="product_pod")

  for book in books:

    # Getting the rating -
    rating_tag = book.find("p", class_="star-rating")
    ratingI = rating_tag["class"][1] # extracts rating tag's class and get the rating
    rating.append(ratingI)

    # Getting the name -
    nameI = book.find("h3").text
    name.append(nameI)

    # Getting the price -
    priceI = book.find("p", class_="price_color").text
    price.append(priceI)

    #Getting the Stock Status -
    stock_tag = book.find("p", class_="instock availability")
    stockI = stock_tag.text.strip()
    is_in_stock.append(stockI)

  d = {
      "name":name,
      "rating":rating,
      "price":price,
      "is_in_stock":is_in_stock
  }
  df = pd.DataFrame(d)
  df_final = pd.concat([df_final, df], ignore_index = True)

df_final
```
