# Performing EDA

##Univariate Analysis

### Categorical Data

i) Countplot (to check frequency distribution among categories)

```python
df['Survived'].value_counts().plot(kind='bar')
```

ii) Pie Chart to check % distribution

```python
df['Pclass'].value_counts().plot(kind='pie')
```

### Numerical Data

a) Histogram

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'], bins=20)
```

b) Distplot

```python
sns.distplot(df['Age']) # gives a probability distribution function, to get probability at a given data point
```

c) Boxplot

```python
sns.boxplot(df['Age'])
```

```python
df['Age'].skew() # 0 means perfectly symmetrical
```
